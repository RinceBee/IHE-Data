# Chapter 5. Sinuoidal fluctuation of river stage

A river level fluctuates according to a sine function given as

$$s = A e^{-a x} sin(\omega t - a x + \theta)$$

Given the envelopes of this fluctuation and show the wave in the aquifer at a number of times using the following variable values:

$kD = 500\,  m2/d$

$A  = 2 \, m$

$\omega = 1/\pi \, d^{-1}$

$S = 0.001$

$\theta = 0.2/\pi$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define the variables and their values
T     = 0.5 # [d] cycle time
omega = 2 * np.pi / T  # [1/d]
eta   = 0.2 * np.pi    # [-]
S     = 0.01           # [-]   
kD    = 1000           # m2/d
A     = 2              # m
a     = np.sqrt(omega / 2 * S / kD)
x     = np.linspace(0, 500, 501)  # m, choose values to compute and show

env1  = +A * np.exp(-a * x)  # top envelope
env2  = -A * np.exp(-a * x)  # bottom envelop


# Setting up a nice plot
plt.title('Sinusoidal fluctation at x=0')
plt.xlabel('$x$ [m]')
plt.ylabel('$s = [h - h_0]$ [m]')
plt.grid()

# plot the envelopes
plt.plot(x, env1, 'b', label='top envelope')
plt.plot(x, env2, 'b', label='botom envelope')

# choose values for time
time = np.arange(0, 0.5, 0.05)

# do for each time, plot a wave
for t in time:
    s = A * np.exp(-a * x) * np.sin(omega * t - a * x + eta)
    plt.plot(x, s, label='t = {:.2f}'.format(t))
   
# finally plot a legend
plt.legend(loc='right')

# make sure the plot is shown
plt.show()
```

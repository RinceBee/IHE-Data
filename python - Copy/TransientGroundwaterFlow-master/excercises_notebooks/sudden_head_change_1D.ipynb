# Sudden change of head at x=0, 1 dimensional flow

The aquifer is considered of constant transmissivity $kD$ and storage coefficient $S$ and to extend from $0 \le x \le \infty$.

The partial differential equation is

$$ kD \frac {\partial^2 s} {\partial x} = S \frac {\partial s} {\partial t}  $$

The solution for a sudden change of head equal to $A [m]$ at $x=0$ and $t = 0$ is

$$ s(x, t) = A \,\mathtt{erfc}(u), \,\,\,\, u=\sqrt{\frac {x^2 S} {4 kD t}} $$

where $\mathtt {erfc} () $ is the so-called complementary error function:

$$ \mathtt {erfc} (z) = \frac 2 {\sqrt {\pi} } \intop _z ^\infty e ^{-y^2}dy $$

And so its derivative is

$$ \frac {d \mathtt{erfc}(z)} {d z} = - \frac 2 {\sqrt {\pi}} e ^{-z^2} $$

Therefore, the discharge equals

$$ Q = -kD \frac {\partial s} {\partial x} = A \sqrt{\frac {kDS} {\pi t}} \mathtt{exp} \left( -\frac {x^2 S} {4 kD t} \right) $$

and for $ x = 0 $

$$ Q_0 = A \sqrt{\frac {kD S} {\pi t}}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erfc
```

```python
kD = 900 # m2/d
S = 0.1 # [-]

A = 1.5 # m

x = np.linspace(0, 1000, 101)
times = [0.1, 1, 2, 4, 8, 16, 32, 64] # days

# plot heads
plt.title("Effect of sudden head change at $x$=0 and $t$=0")
plt.xlabel('x [m]')
plt.ylabel('head change s [m]')
plt.grid()
for t in times:
    u = np.sqrt((x**2 * S)/(4 * kD * t))
    s = A * erfc(u)
    plt.plot(x, s, label='t = {:.0f} d'.format(t))
    
plt.legend()

plt.show()

# plot discharges
plt.title("Discharge due to sudden head change at $x$=0 and $t$=0")
plt.xlabel('x [m]')
plt.ylabel('Discharge Q [m$^2}/d]')
plt.grid()
for t in times:
    u = np.sqrt((x**2 * S)/(4 * kD * t))
    Q = A * np.sqrt((kD * S)/(np.pi * t)) * np.exp(-u**2)
    plt.plot(x, Q, label='t = {:.02f} d'.format(t))
    
plt.legend()

plt.show()
```

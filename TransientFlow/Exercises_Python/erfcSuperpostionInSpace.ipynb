# Superposition in space using the solution

$$ s = \phi - \phi_0 = a \,\, erfc \left( u \right) , \,\,\, u = \sqrt { \frac {x^2 S_y} { 4 kD t } } $$

We solve the case for an aquifer of length $L$ bounded by two parallel surface waters in full contact with the aquifer, after the water levels in both surface water suddenly were raised over a height $a$. The center of the coordiante system is halfway the aquifer.

```python
import numpy as np
import matplotlib.pyplot as plt
from   scipy.special import erfc
from   numpy import pi
```

Aquifer characteristices:

```python
kD = 1200 # m2/d, transmissivity
Sy = 0.2  # -,    storage coefficient, either specific yield or elastic
L  = 300  # m     width of the system (distance between water courses)
```

Boundary conditions

```python
a  = 1   # m, amount of sudden rise of surface water
```

Coordinates and time

```python
x = np.linspace(-L/2,L/2,201)   # 201 points between -L/2 and +L/2
times = np.linspace(0,  5, 10)  # compute for 10 times
times[0]=times[1]/1000          # make first time small but possitive
```

## Solution using

$$ s = a \sum_{i=1}^\infty 
 \left( -1 \right)^\left( i-1 \right) \left(
 erfc \left( \left( \left(i-\frac 1 2\right) L + x \right) \sqrt { \frac {S_y} {4 kD t} } \right) +
 erfc \left( \left( \left(i-\frac 1 2\right) L - x \right) \sqrt { \frac {S_y} {4 kD t} } \right) 
 \right) $$

```python
fig = plt.figure(figsize=(8,5))   # setup plot, generate a figure first
ax  = fig.add_subplot(1,1,1)      # add an axis to the figure
ax.set_ylim(0,1.5*a)              # fix the ylime of axis

for t in times:                   # for each time
    sTotal = np.zeros_like(x)
    for i in np.arange(1,21):     # carry out the summation
        s = a * (-1)**(i-1) * (erfc(((i-0.5)*L+x) * np.sqrt(Sy/ (4 * kD * t))) +
                               erfc(((i-0.5)*L-x) * np.sqrt(Sy/ (4 * kD * t)))
                              )
        sTotal += s      # add partial sum to total
    sTotal = a - sTotal  # changes to the case where the level is reduced from a to zero m
    ax.plot(x,sTotal)    # plot total sum for this time
```

## Solution using instead

$$ s = \frac {4 a} \pi \sum_{j=1}^\infty  \cos \left[ \left( 2 j - 1 \right) \pi \frac k L \right]     \exp \left[ - \left( 2 j - 1 \right) \pi^2 \frac {kD} {L^2 S_y} t \right] $$


```python
T = (L/np.pi)**2 * Sy/kD  # use T as a new variable (is a constant, simplifies formula below)

fig = plt.figure(figsize=(8,5))   # generate figure
ax  = fig.add_subplot(1,1,1)      # add axis to figure
ax.set_ylim(0,1.5*a)              # fix its ylim

for t in times:                   # for each time
    sTotal = np.zeros_like(x)
    for j in range(1,51):         # carry out the summation
        s = 4*a/pi * (-1)**(j-1) / (2*j-1) * cos((2*j-1)*pi*x/L) * np.exp(-(2*j-1)**2 * t/T)
        sTotal += s               # add partial sum to total
    ax.plot(x,sTotal)             # plot final total sum
    
```

#The propagation of a wave of the state in the surface water level on the head in adjacent groundwater.

The solution is

$$ s = s_0 e^{-ax} sin \left( \omega t - a x \right) $$ 

```python
import matplotlib.pylab
%pylab inline
```

```python
kD   = 1200  # m2/d
S    = 0.002 # specific yield
s0   = 1.0   # wave amplitude
omega = 4*pi # tide twice a day 
a = sqrt(S * omega/(2 * kD))
```

```python
x = linspace(0,1000,500)
time = arange(0,1,0.05)
```

```python
def dd(x,t):
  return s0 * exp(-a*x)*sin(omega*t-a*x)
```

```python
for t in time:
  plot(x,dd(x,t))

hold('on')

plot(x, s0*exp(-a*x), linewidth=5); # envelope
plot(x,-s0*exp(-a*x), linewidth=5); # envelope

xlim(0,1000)

xlabel('x [m]'); ylabel('head change s [m]');
title('head as fuction of x from river at different times');
```

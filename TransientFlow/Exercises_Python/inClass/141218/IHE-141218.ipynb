```python
kD = 1200 # m2/d
S  = 0.2  # [-]
x  = linspace(0,10000,200)
time  = linspace(0,250,25) # d
A  = 1 # m
```

```python
omega = 0.5/(2 * pi); # half a day, tide is twice a day 
a = sqrt(omega * S/(2 * kD))
```

```python
def s(x,t):
  return A * exp(-a*x) * sin(omega * t - a * x)
```

```python
for t in time:
    plot(x,s(x,t))
    
ylim(-A,A); xlim(0,10000)
```

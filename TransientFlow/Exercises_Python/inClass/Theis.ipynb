```python
import pylab
from scipy.special import expn
```

First define the aquifer properties and the extraction the well

```python
kD = 900   # [m2/d]
S  = 0.2   # [ - ]   storage coefficient
Q = 1200   # [m3/d]  extraction
```

Let's put in distance and time

```python
r = logspace(0,3,50).reshape(50,1)
t = logspace(-3,1,40)
#print r,t
```

Let's go

```python
s = Q/(4 * pi * kD) * expn(1, S/(4*kD) * outer(r**2,1/t))
#print s
#print s.shape
```

```python
plot(r,-s,'r');

xscale('log')
xlabel('r [m]')
ylabel('drawdown [m]')
title('Theis drawdown')
grid('on')
```

# Transient groundwater flow, analytical solution
## Exercise, dynamic drawdown for `N` wells according to Theis (1935)

The first exercise applied the famous Theis equation given the drawdown `s` due to pumping a well at a constant rate `Q` from `t=0`, in an aquifer of inifnite extent with constant transmissivity `kD` and storage coefficient `S`. Notice the difference between lowercase `s` for drawdown and uppercase `S` for storage coefficient.

In this exercise we will use it to compute the simultaneous drawdown of many wells simultaneously.

This will, of course, be a matter of superpositioin.

```python
from matplotlib.pylab import *  # so that all necessary functions are at hand
from scipy.special import expn  # this is the exponential integral W(u) of Theis
from numpy import zeros
```

expn(1,u) is the exponential integral of Theis' well function as it's known among hydrologists
instead of using expn, we may define it as the well function like this

```python
def W(u):
  return expn(1,u)
```

which is just a matter of taste.

Aquifer properties and flow rate

```python
kD = 900    # m2/d
S  = 0.2    # [-]
x = r_[-1000:1000:5];
y = c_[-1000:1000:5];
```

```python
X = outer(ones(shape(y)),x)
Y = outer(x,ones(shape(x)))
```

```python
t = r_[1,2,4,8.0]
```

Define the wells using a well name + a tuple specifying Q, x and y.

```python
well = {'well1' : (-400, -50, -50),
        'well2' : (-400, +50, -50),
        'well3' : (-400, +50, +50),
        'well4' : (-400, -50, +50)}
```

Now carry out the superposition by looping over the wells and adding the drawdowns for all points in the grid.

```python
s = np.zeros(shape(X))  # initilize drawdown to all zeros

for w in well:
    r = sqrt((X-well[w][1])**2 +(Y-well[w][2])**2)          # r between well and grid points
    s += well[w][0]/(4 * pi * kD) * W(r**2/(4*kD*t[0]))     # s drawdown accumlating
```

Contour the drawdowns, using full color plot and colorbar

```python
contourf(X,Y,s,10)  # full color plot
colorbar();
xlim(-200,200)      # zoom in on x
ylim(-200,200)      # zoom in on y
```

Finished. You can now compute the drawdown for any time and any number of wells.

# Transient groundwater flow, analytical solution
## Exercise, dynamic drawdown according to Theis (1935)

The famous Theis equation given the drawdown `s` due to pumping a well at a constant rate `Q` from `t=0`, in an aquifer of inifnite extent with constant transmissivity `kD` and storage coefficient `S`. Notice the difference between lowercase `s` for drawdown and uppercase `S` for storage coefficient.

The formula is valid for constant aquifer parameters, i.e. transmissivity and storage coefficient and no head boundary conditions. The only condition is the extraction from the well, starting at `t=0`.


```python
from matplotlib.pylab import *  # so that all necessary functions are at hand
from scipy.special import expn  # this is the exponential integral W(u) of Theis
```

expn(1,u) is the exponential integral of Theis' well function as it's known among hydrologists
instead of using expn, we may define it as the well function like this

```python
def W(u):
  return expn(1,u)
```

which is just a matter of taste.

Aquifer properties and flow rate

```python
kD = 900    # m2/d
S  = 0.2    # [-]
Q = 1200   # m3/d
```

Definte coordinates, in this case distrances `r` from the center of the well and times `t`. We use logarithmically increasing distances in both time and space to bridge the manu orders of magnitude for both `r` and `t`.

```python
r = logspace(0,3,50)[newaxis].T   # fifty points between 1 and 1000 m (10**0 and 10**3 m)
t = logspace(-3,np.log10(3),25)   # times on the horizontal axis
```

**Theis groundwater solution:** Note the outer product to multiply the column vector `r` with the row vector `t` to get a full matrix in which every element is the product of the corresponding `r` and `t` values.

```python
s = Q/(4 * pi * kD) * W(S/(4*kD) * outer(r.T**2,1/t))
```

**Plotting:** We plot drawdown versus distance for different distances. Notice the dimension of both `r` and `s`. The plot will use `r` as the x-axis and will plot each column (which corresponds to a particular time) as a separate line.

```python
plot(r,s);

xscale('log')         # r axis on logarithmic

# embellishment of graph
xlabel('log(r) [m]')
ylabel('drawdown [m]')
title('First attempt to make a groundwater solution in Python');
grid('on')

# Because we like to see the drawdown being drawn downward, we invert the yaxis.
ax = gca()            # gca() is function that returns handle to current axis
ax.invert_yaxis()     # use method invert_yaxis() to actually invert the yaxis
```

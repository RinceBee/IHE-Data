Theis solution for transient well in infinite aquifer with constant transmissivity and storavity, i.e constant values of KD and S.
The this solution is given by,
$$ s= \frac {Q}{ 4 \pi kD t} W(u), u = \frac{r^2 S}{ 4 kD t } $$

and the function W can be defined as, 
$$expint(u)=E_1 (u) = \intop_u^\infty \frac{e^{-y}}{y} dy $$

the graph W(u) vs 1/u is called thies type curve.
Lets define kD and S


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sp
```

```python

kD= 900 # m2/day
S = 0.25 # [-]
u=np.logspace (-6, 1, 101) # in 101 steps u= 10e-6 to 10
plt.plot( 1/u, sp.exp1(u)) # 1/u as selfdefined, exp1 is W(u) as defined by defination 
plt.title('Theis Type curve')
plt.xlabel('1/u------->')
plt.ylabel('W(u)')
plt.xscale('log')
plt.yscale('log')
plt.grid()
plt.show()
```
